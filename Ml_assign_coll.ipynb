R-squared represents the proportion of the variance in the dependent variable that is explained by the independent variables in a regression model. It indicates how well the model fits the data.

3. What is multicollinearity, and how does it impact regression

Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. This means one predictor variable can be linearly predicted from the others, leading to redundancy in the model.

11. What is the role of homoscedasticity in linear regression

Homoscedasticity means that the variance of residuals (errors) remains constant across all levels of the independent variable(s). It is a key assumption in linear regression.

14. What is heteroscedasticity, and why is it a problem

Heteroscedasticity refers to a situation in regression analysis where the variance of errors (residuals) is not constant across all levels of the independent variable. This means that as the predictor values change, the spread of residuals increases or decreases unevenly rather than remaining consistent.

1️. Violates Assumptions → Linear regression assumes constant variance (homoscedasticity). If violated, predictions may be unreliable.
2️. Biased Standard Errors → Can lead to incorrect hypothesis testing (e.g., misleading p-values and confidence intervals).
3️. Affects Model Interpretation → Reduces the accuracy of coefficients and weakens statistical inference.

15. How can interaction terms enhance a regression model's predictive power

In regression, interaction terms capture the combined effect of two independent variables when their influence on the dependent variable is not purely additive but instead depends on each other.
 Interaction terms capture hidden relationships between features, improving the regression model’s predictive power.
